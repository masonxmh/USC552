# 4. ISLR, 6.6.5

#### (a)
We need to minimize the below equation in ridge regression: <br>

$\sum_{i=1}^{n}(y_{i} - \beta_0 - \sum_{j=1}^{p}\beta_jx_{j})^2 + \lambda \sum_{j=1}^{p}\beta_j^2$

Since $n=2, p=2, \hat{\beta}_{0} = 0$ rewrite the equation


$\min( (y_1 - \hat{\beta}_1x_{11} - \hat{\beta}_2x_{12})^2 + (y_2 - \hat{\beta}_1x_{21} - \hat{\beta}_2x_{22})^2 + \lambda (\hat{\beta}_1^2 + \hat{\beta}_2^2) )$

#### (b)

$f(\hat{\beta}_1, \hat{\beta}_2) =  (y_1 - \hat{\beta}_1x_{11} - \hat{\beta}_2x_{12})^2 + (y_2 - \hat{\beta}_1x_{21} - \hat{\beta}_2x_{22})^2 + \lambda (\hat{\beta}_1^2 + \hat{\beta}_2^2)$  &nbsp;&nbsp; eqn(1)

$x_{11} = x_{12}, x_{21} = x_{22}$

$y_{1} + y_{2} = 0, x_{11}+x_{21} = 0, x_{12}+x_{22} = 0 $

$y_{2} = - y_{1}$

$x_{21} = x_{22} = - x_{11}$

Rewrite eqn(1):

$f(\hat{\beta}_1, \hat{\beta}_2) =  (y_1 - \hat{\beta}_1x_{11} - \hat{\beta}_2x_{11})^2 + (-y_1 + \hat{\beta}_1x_{11} + \hat{\beta}_2x_{11})^2 + \lambda (\hat{\beta}_1^2 + \hat{\beta}_2^2)$

$f(\hat{\beta}_1, \hat{\beta}_2) =  2(y_1 - \hat{\beta}_1x_{11} - \hat{\beta}_2x_{11})^2 + \lambda (\hat{\beta}_1^2 + \hat{\beta}_2^2)$  

$f(\hat{\beta}_1, \hat{\beta}_2) =  2y_{1}^2 - 4y_{1}x_{11}\hat{\beta}_1 - 4y_{1}x_{11}\hat{\beta}_2 - 4x_{11}^2\hat{\beta}_1\hat{\beta}_2 + 2x_{11}^2\hat{\beta}_1^2 + 2x_{11}^2\hat{\beta}_2^2 + \lambda (\hat{\beta}_1^2 + \hat{\beta}_2^2)$

In order to minimize $f(\hat{\beta}_1, \hat{\beta}_2)$ , we need to take the deriviatative of this function respect to $\hat{\beta}_1$ and $\hat{\beta}_2$ equal to 0.

Then we can obtain:

$\hat{\beta}_1 = \frac{2x_{11}y_1 - 2x_{11}^2\hat{\beta}_2}{2x_{11}^2 + \lambda}$

$\hat{\beta}_2 = \frac{2x_{11}y_1 - 2x_{11}^2\hat{\beta}_1}{2x_{11}^2 + \lambda}$

Since $\hat{\beta}_1$ and $\hat{\beta}_2$ are independent from $x_{11}$ and $y_1$

Then the equation have to satisfy $\hat{\beta}_1$ = $\hat{\beta}_2$

#### (c)
Lasso expression:

$\sum_{i=1}^{n}(y_{i} - \beta_0 - \sum_{j=1}^{p}\beta_jx_{j})^2 + \lambda \sum_{j=1}^{p}|\beta_j|$

Then plug in the conditions:

$x_{11} = x_{12}, x_{21} = x_{22}$

$y_{1} + y_{2} = 0, x_{11}+x_{21} = 0, x_{12}+x_{22} = 0 $

$y_{2} = - y_{1}$

$x_{21} = x_{22} = - x_{11}$

We can obtain:

$2(y_1 - \hat{\beta}_1x_{11} - \hat{\beta}_2x_{11})^2 + \lambda (| \hat{\beta}_1 | + | \hat{\beta}_2 |)$

#### (d)
Based on (c) we know that  $2(y_1 - \hat{\beta}_1x_{11} - \hat{\beta}_2x_{11})^2 \geq 0 $ when $| \hat{\beta}_1 | + | \hat{\beta}_2 | \leq s $

Then we can have:

$\hat{\beta}_1 + \hat{\beta}_2  = \frac{y_1}{x_{11}}$

So that $\hat{\beta}_1$ and $\hat{\beta}_2$ won't be unique

# 6. ISLR 9.7.3
ref: https://scikit-learn.org/stable/auto_examples/svm/plot_separating_hyperplane.html

```python
import matplotlib.pyplot as plt
from sklearn import preprocessing
import numpy as np
from sklearn import svm
```

### (a) Sketch the observations

```python
y_ori = ['red', 'red', 'red', 'red', 'blue', 'blue', 'blue']
X = np.array([[3,4], [2,2], [4,4], [1,4], [2,1], [4,3],[4,1]])
le = preprocessing.LabelEncoder()
y = le.fit_transform(y_ori)

plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_ori)
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid()
plt.show()
```

### (b) Sketch the optimal separating hyperplane, and provide the equation for this hyperplane (of the form (9.1)).

```python
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_ori)
# fit the model, don't regularize for illustration purposes
clf = svm.SVC(kernel="linear", C=1000)
clf.fit(X, y)
# plot the decision function
ax = plt.gca()
xlim = (ax.get_xlim())
ylim = (ax.get_ylim())
# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)
# plot decision boundary and margins
ax.contour(XX, YY, Z, colors=["white", "k", "white"], levels=[-1, 0, 1], alpha=0.5, linestyles=["--", "-", "--"])
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid()
plt.show()
```

The solid line is the hyperplane

```python
intercept = clf.intercept_
coef = clf.coef_
print("Intercept is: ", intercept)
print("Coefficient is: ,", coef)
```

So the result can be written as:

$0.99970703 - 1.99941406X_1 + 1.99941406X_2 = 0$

simplify:

$0.5 -X_1 + X_2 = 0$

If we calculate the hyperplane by hand, we can find this hyperplane line by observation with a margin of $\frac{\sqrt2}{4}$. Then we can obtain the equation the same as simplified sklearn result

$0.5 -X_1 + X_2 = 0$

```python
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_ori)
# fit the model, don't regularize for illustration purposes
clf = svm.SVC(kernel="linear", C=1000)
clf.fit(X, y)
# plot the decision function
ax = plt.gca()
xlim = (ax.get_xlim())
ylim = (ax.get_ylim())
# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)
# plot decision boundary and margins
ax.contour(XX, YY, Z, colors="k", levels=[-1, 0, 1], alpha=0.5, linestyles=["--", "-", "--"])
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid()

# plot support vectors
ax.scatter(
    clf.support_vectors_[:, 0],
    clf.support_vectors_[:, 1],
    s=200,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
plt.show()
```
